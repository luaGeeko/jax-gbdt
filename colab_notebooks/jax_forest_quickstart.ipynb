```python
import sys
import os
# Add repo root dynamically
repo_path = os.path.abspath("..")
sys.path.append(repo_path)
```

```python
from jax_forest.api import JAXForestRegressor
import numpy as np
```

```python
# 1. Initialize the compiler with an existing XGBoost model
repo_path = os.path.abspath("..")  # your repo root
trained_model_path = os.path.join(repo_path, "checkpoints/xgboost_model/xgboost_model.json")
model = JAXForestRegressor(
    model_path=trained_model_path, 
    method="soft_dense"
)

# 2. Compile it for the hardware (using a dummy batch of 500k samples)
X_data = np.random.rand(500000, 8)
print("Compiling...")
model.compile(X_data)

# 3. Blazing fast inference in production
X_real = np.random.rand(500000, 8)
predictions = model.predict(X_real)
print("Predictions:", predictions.shape)
```
